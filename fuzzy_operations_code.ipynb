```python
 import numpy as np
 import matplotlib.pyplot as plt
 import skfuzzy as fuzz
 from skfuzzy import control as ctrl
```

```python
x=np.arange(0,11,1)
```

```python
def fuzzy_set_A(x):
  return np.maximum(0,np.minimum((x-2)/3,(8-x)/3))
```

```python
def fuzzy_set_B(x):
  return np.maximum(0,np.minimum((x-4)/3,(10-x)/3))
```

```python

# Calculate the membership values for fuzzy sets A and B
membership_A = fuzzy_set_A(x)
membership_B = fuzzy_set_B(x)
```

```python
# Plot fuzzy set B
plt.figure(figsize=(8, 4))
plt.plot(x, membership_B, 'g', label='Fuzzy Set B')
plt.title('Fuzzy Set B')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership')
plt.legend()
plt.grid()
plt.show()
```

```python
# Calculate and plot the intersection of fuzzy sets A and B
intersection = np.minimum(membership_A, membership_B)

plt.figure(figsize=(8, 4))
plt.plot(x, intersection, 'r', label='Intersection (A ∩ B)')
plt.title('Fuzzy Set Intersection (A ∩ B)')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership')
plt.legend()
plt.grid()

plt.show()

```

```python
# Union of a Fuzzy Set
import numpy as np
import matplotlib.pyplot as plt

# Define the universe of discourse (X-axis)
x = np.arange(0, 11, 1)

# Define the triangular fuzzy sets A and B as membership functions
def fuzzy_set_A(x):
    return np.maximum(0, np.minimum((x - 2) / 3, (8 - x) / 3))

def fuzzy_set_B(x):
    return np.maximum(0, np.minimum((x - 4) / 3, (10 - x) / 3))

# Calculate the union of fuzzy sets A and B
union = np.maximum(fuzzy_set_A(x), fuzzy_set_B(x))

# Plot fuzzy sets A and B
plt.plot(x, fuzzy_set_A(x), 'b', label='Fuzzy Set A')
plt.plot(x, fuzzy_set_B(x), 'g', label='Fuzzy Set B')

# Plot the union of fuzzy sets A and B
plt.plot(x, union, 'r', label='Union (A ∪ B)')

plt.title('Fuzzy Set Union (A ∪ B)')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership')
plt.legend()
plt.grid()

plt.show()
```

```python
# Complement a Fuzzy Set
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Create the universe of discourse (universal set)
x = np.arange(0, 11, 1)  # A universe of discourse from 0 to 10

# Create a triangular fuzzy set
set_A = fuzz.trimf(x, [0, 3, 6])  # A triangular fuzzy set

# Plot the fuzzy set
plt.figure(figsize=(8, 4))
plt.plot(x, set_A, 'b', linewidth=2, label='Set A')
plt.title('Fuzzy Set')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership')
plt.legend()
plt.grid()

# Perform fuzzy complement (NOT) operation
set_complement_A = 1 - set_A  # complement = 1 - μ_A(x)

# Plot the complemented fuzzy set
plt.figure(figsize=(8, 4))
plt.plot(x, set_complement_A, 'c', linewidth=2, label='Complement (NOT) of Set A')
plt.title('Fuzzy Complement (NOT) Operation')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership')
plt.legend()
plt.grid()

# Show the plots
plt.show()
```
